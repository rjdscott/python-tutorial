```python
def get_manhattan_avg(file_input_path, file_output):
    df = pd.read_csv(file_input_path)
    df[df['neighbourhood_group']=='Manhattan'].groupby('neighbourhood').mean().sort_values('price',ascending=False).to_csv(file_output)
    
```
